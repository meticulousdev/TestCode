```python
import pandas as pd
```

```python
day_to_sec = 86400
```

```python
idx = pd.date_range('2023-01-01', periods=day_to_sec, freq='S')
print(idx)
```

```python
data_columns = ['data01', 'data02', 'data03', 'data04']

data_example = pd.DataFrame([[i, i * 2, i * i, 0] for i in range(idx.shape[0])], index=idx, columns=data_columns)
print(data_example)
```

```python
data_hourly_mean = pd.DataFrame()
data_hourly_mean[data_columns] = data_example[data_columns].resample('60T').mean()
print(data_hourly_mean)
```

```python
working_hour = pd.DataFrame()
working_hour['start'] = ['2023-01-01 09:00:00']
working_hour['end'] = ['2023-01-01 18:00:00']

working_hour['start'] = pd.to_datetime(working_hour['start'], errors='coerce')
working_hour['end'] = pd.to_datetime(working_hour['end'], errors='coerce')

working_hour['net'] = working_hour['end'] - working_hour['start']
print(working_hour)
```
