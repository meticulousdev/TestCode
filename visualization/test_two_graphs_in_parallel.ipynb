```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_columns = ['data01', 'data02']
index = ['AAAA', 'BBBB', 'CCCC', 'DDDD', 'EEEE', 'FFFF', 'GGGG']

data_example = pd.DataFrame([[np.sqrt(i+1)*(-1)**(i+1), 1/(i+2)*(-1)**i] for i in range(7)], 
                            index=index, 
                            columns=data_columns)
print(data_example)
```

```python
plt.subplot(1, 2, 1)
plt.barh(index, data_example['data01'], color='red')
plt.xlim([-3, 3])
plt.xlabel('data01')

plt.subplot(1, 2, 2)
plt.barh(index, data_example['data02'], color='blue')
plt.xlim([-1, 1])
plt.xlabel('data02')

plt.show()
```
